```python
from sklearn.datasets import fetch_openml
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

### 1. Load the dataset and conduct any necessary preprocessing, such as normalizing the data.

```python
mnist = fetch_openml("Fashion-MNIST", version=1, cache=True)
```

```python
indices = np.random.choice(70000, 10000)
X = (
    mnist.data[indices] / 255.0
)  # Divide by 255 to normalize for RGB (Red, Green, Blue) in image data
y = mnist.target[indices]

print(X.shape, y.shape)
```

### Apply t-SNE to the data by setting n_components=2.

```python
tsne = TSNE(n_components=2, verbose=1, n_jobs=-1)
components = tsne.fit_transform(X)
```

### Using the two-dimensional t-SNE representation, draw a graph of the data by coloring and labeling the data points as you did in the checkpoint.

```python
plt.figure(figsize=(10, 5))
df = pd.DataFrame(components)
df["target"] = y.astype(int)

plt.scatter(x=df[0], y=df[1], c=df.target)
plt.colorbar()
```

```python
tsne = TSNE(n_components=2, verbose=1, n_jobs=-1, perplexity=20)
components = tsne.fit_transform(X)
plt.figure(figsize=(10, 5))
df = pd.DataFrame(components)
df["target"] = y.astype(int)

plt.scatter(x=df[0], y=df[1], c=df.target)
plt.colorbar()
```

```python
tsne = TSNE(n_components=2, verbose=1, n_jobs=-1, perplexity=40)
components = tsne.fit_transform(X)
plt.figure(figsize=(10, 5))
df = pd.DataFrame(components)
df["target"] = y.astype(int)

plt.scatter(x=df[0], y=df[1], c=df.target)
plt.colorbar()
```
